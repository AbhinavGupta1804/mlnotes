```python
import pandas as pd
messages=pd.read_csv('SpamClassifier-master/smsspamcollection/SMSSpamCollection',
                    sep='\t',names=["label","message"])

# messages.shape        #(5572, 2)

from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
import re
import nltk
nltk.download('stopwords')
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review]
    review = ' '.join(review)
    corpus.append(review)
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus, messages['message']) if i<1]    
#[[i, j, k] ...]: Stores the results in a list of [length, processed text, original text].
#with the help of this , we will be able to find 3 deleted rows
```

#### This list comprehension is filtering out empty messages from the corpus.

map(len, corpus): Gets the length of each preprocessed text in corpus.

zip(...): Combines:

Length of each preprocessed text (i),
Preprocessed text (j from corpus),
Original message (k from messages['message']).

if i < 1: Keeps only those where i (length of j) is less than 1 (empty).

Purpose:
It helps identify empty or fully removed texts after preprocessing.

### simple_preprocess(sent):
Converts the sentence into lowercase.
Removes punctuation.
Tokenizes it into words.

```python
#this is an alternative method of preprocessing and cleaning
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)   #Tokenizes the sentence into smaller sentences using sent_tokenize().
    for sent in sent_token:                   #Iterates through each split sentence 
        words.append(simple_preprocess(sent))
words
#The list words now contains tokenized words from all sentences. 
```

#### By default, gensim's Word2Vec trains on 100 dimensions if you dont specify it explicitly.

### How are dimensions automatically learned in Word2Vec?
 The Word2Vec model learns values for these dimensions automatically based on the training data.

2 Start with random numbers => Initially, every word is assigned a random vector of 100 numbers.

3 Training phase: Learning from context

The model scans sentences and sees which words appear together.
It adjusts the vectors so that words used in similar contexts get similar vectors.
.
 Why is this "automatic"?
We dont tell the model what each of the 100 numbers means.
### Instead, the model learns patterns from the data and automatically assigns the best values to these 100 dimensions.

 Simple Analogy
Think of a 100-question personality test .

Each person answers differently, and we get a 100D profile for each.
But we dont predefine what each question meanswe just use patterns in answers to group similar people together.

```python
import numpy as np
# def avg_word2vec(doc):
#     # remove out-of-vocabulary words
#     #sent = [word for word in doc if word in model.wv.index_to_key]
#     #print(sent)
#                     #vector with 100d
#     return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
#                 #or [np.zeros(len(model.wv.index_to_key))], axis=0) 
def avg_word2vec(doc):
    valid_words = [word for word in doc if word in model.wv.index_to_key]
    if len(valid_words) == 0:  
        return np.zeros(100)  # Return a 100D zero vector for empty cases
    return np.mean([model.wv[word] for word in valid_words], axis=0)


                
                #np.zeros(len(model.wv.index_to_key)) ensures that if no valid word embeddings exist,
                #  it returns a vector of zeros instead.
```

tqdm shows a progress bar when a loop is running, so you can see how much is done and how much is left. 

This loops through each list of words (sentence) in words, applies the avg_word2vec function to get its average Word2Vec embedding, and stores the result in X.

```python
!pip install tqdm
from tqdm import tqdm
#apply for the entire sentences
X=[]                                    #The list words now contains tokenized words from all sentences. 
for i in tqdm(range(len(words))):  
    X.append(avg_word2vec(words[i]))
len(X)                                  #5569 tokenized sentences
X                   # this is a list of NumPy array
    
```

```python
##independent Features
X_new=np.array(X)
# messages.shape     #(5572, 2)

# X[1]
# X_new.shape    # (5569, 100)     # 3 rows only had expressions & numbers , so they got deleted
# X_new[0].shape  #(100,)
```

### map(lambda x: len(x) > 0, corpus):
Checks if each preprocessed message in corpus has a length greater than 0 (i.e., it's not empty).
Returns a list of True/False values.
### messages[...]:
Filters rows in messages where the corresponding preprocessed text in corpus is not empty.

```python
## Dependent Features
## Output Features
y = messages[list(map(lambda x: len(x)>0 ,corpus))]
y=pd.get_dummies(y['label'])
y=y.iloc[:,0].values                  #selecting first column 
y.shape  #(5569,)
```

1  Keep 1 row.

-1  Automatically adjust the number of columns (so it remains 100).

Shape changes from (100,)  (1, 100).

This is useful for machine learning models that expect 2D inputs instead of 1D vectors.

```python
X[0].reshape(1,-1).shape   #(1, 100)
```

ignore_index=True ensures that the index gets reset instead of keeping the original indices.

```python
## this is the final independent features
df=pd.DataFrame()                             #reshaping all arrays                                     
    
df = pd.concat([pd.DataFrame(X[i].reshape(1, -1)) for i in range(len(X))], ignore_index=True)
df['Output']=y
df.head()
```

```python
df.dropna(inplace=True)                        #row wise deletion
# df.isnull().sum()    #0 
# df.shape                 #(5569, 101)
```

```python
## Independent Feature
X=df.drop(columns = ['Output'])
X.isnull().sum()                                    #0
y=df['Output']
# X.shape              #(5569, 100)
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
X_train.head()
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test,y_pred))

```

```python
print(classification_report(y_test,y_pred))
```
